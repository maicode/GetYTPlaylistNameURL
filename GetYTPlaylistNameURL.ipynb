```python
# If you don't have the api client, uncommment the line below
# pip install google-api-python-client

from apiclient.discovery import build
```

```python
youtube = build('youtube', 'v3', developerKey = api_key)
```

```python
# Results are limited to 50 per query. This function gets every video in a playlist.
def get_playlist_videos():

    videos = []
    next_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id,
                                              part = 'snippet',
                                              maxResults = 50,
                                              pageToken = next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
            
    return videos
```

Change playlist_id to the playlist you want to get data from.

```python
videos = get_playlist_videos()
```

```python
import csv

f = open('_playlist.csv', 'w')

for i in range(len(videos)):
    title = videos[i]['snippet']['title']
    titleWithoutCommas = title.replace(',', '$')
    imgUrl = videos[i]['snippet']['thumbnails']['default']['url']
    # This is the whole ytimg.com url
    # (eg. 'https://i.ytimg.com/vi/3wgcahcerBM/mqdefault.jpg')
    # ...because there is no videoId in snippets from playlists
    videoId = imgUrl[23:-12] 
    # Takes the videoId part of url (eg.'3wgcahcerBM')
    # If the code breaks, it was probably videoId.
    fullUrl = "https://www.youtube.com/watch?v=" + videoId
    
    f.write(titleWithoutCommas)
    f.write(', ')
    f.write(fullUrl)
    f.write('\n')

    print(titleWithoutCommas)
    print(fullUrl)
        
f.close()
```
